## Required Imports for the Project




```python
# imports
import pandas as pd
import os
```

```python
# import the dataset from kaggle
# dataset URL -> https://www.kaggle.com/datasets/ahmadshogran/the-titanic
import kagglehub
path = kagglehub.dataset_download("ahmadshogran/the-titanic")
```

## Loading the Dataset

### Dataset Description

- **Total Passengers:** 891
- **Features:** 12 (including target `Survived`)
- **Target Variable:** `Survived` (0 = No, 1 = Yes)
- **Key Features:** Passenger class (`Pclass`), Sex, Age, Fare, Cabin, Embarked
- **Missing Data:** Present in `Age`, `Cabin`, and `Embarked`

```python
# Load the Titanic dataset
print(os.listdir(path))
data = pd.read_csv(path + '/titanic.csv')
display(data)
```

## Handle Missing Values

```python
# find the number of NaNs in the dataset
num_nans = data.isna().sum().sum()
print(f'Number of missing values in the dataset: {num_nans}')

# find the columns that contains the NaNs
col_nan = data.isna().sum()
display(col_nan)
```

```python
# Handle missing values in 'Age' with median
data['Age'] = data['Age'].fillna(data['Age'].median())

# Fill missing 'Embarked' with mode
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# Drop 'Cabin' because it has a lot of NaN values
data = data.drop('Cabin', axis= 1)

# Display the cleaned DataFrame without missing values
display(data)
```

```python
# Check the results after handling missing values
num_nans = data.isna().sum().sum()
print(f'Number of missing values in the dataset: {num_nans}')
```

## Remove Duplicates

```python
# Remove Duplicates
# Find the number of duplicates in the dataset and print it
duplicates = data.duplicated().sum()
print(f'Number of Duplicates before removing it -> {duplicates}')

# Remove duplicates
data = data.drop_duplicates(keep='first')

# Display the DataFrame
print('Dataset after removing Duplicated rows:')
display(data)
```

## Convert data types

```python
# display the datatype for each column
data.dtypes
```

```python
# Convert some data types for some columns
data['Survived'] = data['Survived'].astype('bool')
data['Pclass'] = data['Pclass'].astype('category')
data['Sex'] = data['Sex'].astype('category')
data['Embarked'] = data['Embarked'].astype('category')
data['Age'] = data['Age'].astype('int64')

# Standardize values and ensure correct data types
# standardize Sex to lowercase 'male'/'female' and format Fare to two decimal places.
data['Sex'] = data['Sex'].str.lower()
data['Fare'] = data['Fare'].round(2).astype(float)
display(data)
```

## Rename columns

```python
# renaming for some columns in the dataset
data.rename(columns={'PassengerId':'Passenger_Id', 'Pclass':'P_Class', 'SipSp': 'Sibling_Num'}, inplace=True)
display(data)
```

## Save Cleaned Dataset

```python
#save the final cleaned dataset
data.to_csv("titanic_cleaned.csv", index=False)
print("\n'titanic_cleaned.csv' Saved")
```
